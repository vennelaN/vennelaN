```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python

#Random Distribution

#Generate a 1-D array containing 100 values, where each value has to be 3, 5, 7 or 9.
#The probability for the value to be 3 is set to be 0.1
#The probability for the value to be 5 is set to be 0.3
#The probability for the value to be 7 is set to be 0.6
#The probability for the value to be 9 is set to be 0

random_distribution = np.random.choice([3,5,7,9],p=[0.1,0.3,0.6,0],size = (100))
print(random_distribution)
#here if you observe 7 has. the highest probability i.e 0.6 that is the reason you can see that 7 ocuurs most
#and the probability of 9 is 0 ,that is why 9 is not presnt in. the output. 
#and also the total probability 0.1+0.3+0.6+0 = 1
print('*****************************************************************')

#Same example as above, but return a 2-D array with 3 rows, each containing 5 values.
random_distribution = np.random.choice([3,5,7,9],p=[0.1,0.3,0.6,0],size=(3,4))
print(random_distribution)
print('*****************************************************************')
```

```python
#Normal (Gaussian) Distribution
#Use the random.normal() method to get a Normal Data Distribution.
#Generate a random normal distribution of size 2x3:

normal_dist = np.random.normal(size=(2,3))
print(normal_dist)
print('*****************************************************************')

#Generate a random normal distribution of size 2x3 with mean at 1 and standard deviation of 2:
normal_dist2 = np.random.normal(loc=1,scale=2,size=(2,3))
print(normal_dist2)

#Visualization of Normal Distribution
sns.distplot(np.random.normal(size=1000))
plt.show()
```

```python
#Binomial Distribution
#Binomial Distribution is a Discrete Distribution.
#Given 10 trials for coin toss generate 10 data points:

binomial_dist = np.random.binomial(n=10,p=0.5,size =10)
print(binomial_dist)

#Visualization of Binomial Distribution

sns.distplot(binomial_dist)
plt.show()
```
